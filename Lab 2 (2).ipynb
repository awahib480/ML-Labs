# Cleaning
- No negative markings
- Out-of-range scores
- No duplicates
- Change datatype as needed
- No missing values

```python
# checking null values
df.isna().sum()
```

```python
# seeing the place of missing values
df[df['quiz1'].isna()]
df[df['quiz2'].isna()]
df[df['assignment'].isna()]
df[df['final'].isna()]
```

```python
# filling the missing values (with mean of column)
df['quiz1'] = df['quiz1'].fillna(df['quiz1'].mean())
df['quiz2'] = df['quiz2'].fillna(df['quiz2'].mean())
df['assignment'] = df['assignment'].fillna(df['assignment'].mean())
df['final'] = df['final'].fillna(df['final'].mean())
```

```python
#verified
df.isna().sum()
```

```python
# checking for negative value
df['final'] < 0 
```

```python
# replacing negative value with mean
df.loc[5,'final'] = df['final'].mean()
```

```python
# checking for duplicates
df.duplicated().sum()
```
