## Feature Engineering

```python
# Making a new feature 'total_sales' using price and quantity features
df['total_sales'] = df['price'] * df['quantity']
```

```python
# creating a new feature 'sales_category' to segment customers into 4 classes for insights
df['sales_category'] = pd.cut(df['total_sales'], 
                              bins=[0,20000,50000,100000,float('inf')],
                              labels=['Low','Medium','High','Very High'])
```

```python
df['discount'] = df['category'].map({
    'Electronics': 0.05,
    'Accessories': 0.10,
    'Furniture': 0.04
    })
```

```python
# creating 'final price' column after applying discount
df['final_sales'] = df['total_sales'] * (1-df['discount'])
```

```python
df['month'] = df['order_date'].dt.month
df['weekday'] = df['order_date'].dt.day_name
```
