```python
import pandas as pd
```

```python
# Series
deaths = pd.Series(
    [517204, 156938, 254263, 86122, 122705, 86332, 
     69142, 97507, 28503, 70601],)

# DataFrame
data = {
    "Location": ["United States","India","Brazil","Russia",
                 "United Kingdom","France","Spain","Italy",
                 "Turkey","Germany"],
    "Cases": [28833039,11079979,10517232,4246079,4170519,
              3736016,3188553,2907825,2693164,2444169],
    "Deaths": [517204,156938,254263,86122,122705,86332,
               69142,97507,28503,70601],
    "Recovered": [None,10763451,9386440,3811797,None,
                  None,None,2398352,2565723,2242767]
}
df = pd.DataFrame(data)
df
```

```python
df['Status'] = pd.cut(df['Cases'],
                      bins=[0, 3000000, 4000000, float('inf')],
                      labels=['Low', 'Medium', 'High'])
df['Status']
```

```python
deaths.sort_values(ascending=True)
```

```python
df.sort_values(by="Deaths")
```

```python
print('Total Deaths:', deaths.sum())
print('Min deaths:', deaths.min())
print('Max deaths:', deaths.max())
```

```python
print('Total Recovered:', int(df['Recovered'].sum()))
print('Total Cases:', df['Cases'].sum())      
```

```python
mean = df['Recovered'].mean()
df['Recovered'] = df['Recovered'].fillna(mean)
print(df)

median = df['Recovered'].median()
df['Recovered'] = df['Recovered'].fillna(median)
print(df)
```
